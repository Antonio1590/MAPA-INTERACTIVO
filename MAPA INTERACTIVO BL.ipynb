```python
data = {
    'Sede': ['Sede Bogotá', 'Sede Medellín', 'Sede Cali'],
    'Ciudad': ['Bogotá', 'Medellín', 'Cali'],
    'Latitud': [4.60971, 6.2442, 3.4516],
    'Longitud': [-74.08175, -75.5812, -76.532],
    'Colaboradores': [500, 300, 200],
    'Pisos': [10, 8, 5],
    'Horario': ['8 AM - 6 PM', '9 AM - 5 PM', '8 AM - 6 PM']
}
sedes_df = pd.DataFrame(data)
```

```python
mapa = folium.Map(location=[4.570868, -74.297333], zoom_start=6)
```

```python
for i, row in sedes_df.iterrows():
    folium.Marker(
        location=[row['Latitud'], row['Longitud']],
        popup=f"Sede: {row['Sede']}<br>Colaboradores: {row['Colaboradores']}<br>Pisos: {row['Pisos']}<br>Horario: {row['Horario']}",
        tooltip=row['Sede']
    ).add_to(mapa)
```

```python
mapa
```

```python
def zoom_sede(nombre_sede):
    sede = sedes_df[sedes_df['Sede'] == nombre_sede]
    if not sede.empty:
        lat, lon = sede.iloc[0]['Latitud'], sede.iloc[0]['Longitud']
        folium.Map(location=[lat, lon], zoom_start=15).add_child(folium.Marker(location=[lat, lon], popup=nombre_sede))
    else:
        print("Sede no encontrada")
```

```python
mapa.save("mapa_sedes.html")
```
